# CREANDO GRAFICOS CON POO

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5,11)
y = x**2
```

# PLOT USANDO PROGRAMACIÓN ORIENTADA A OBJETOS

## CLASE PLT.FIGURE()
plt.figure() es una función que crea una instancia de la clase matplotlib.figure.Figure.

# FORMA ANTIGUA DE CREAR UN OBJETO DE FIGURE

```python
from matplotlib.figure import Figure
fig_old = Figure()
type(fig_old)
```

# FORMA MODERNA DE CREAR UN OBJETO DE LA CLASE FIGURE

```python
fig = plt.figure()
type(fig)
```

# USANDO LOS METODOS DE LA CLASE FIGURE

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x,y,'r')
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y,'b')
fig.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.15,0.60,0.4,0.3])

axes1.plot(x,y,'b')
axes2.plot(y,x,'r')

axes2.set_facecolor('gray')

fig.show()
```

# CREANDO VARIOS GRAFICOS CON POO Y SUBPLOTS

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y,'r--')
axes[1].hist(y)

plt.show()
```

# CREANDO UNA TUPLA DE AXES

```python
fig,(ax1,ax2) = plt.subplots(nrows=1,ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'g')

plt.show()
```

# CREANDO UNA ARRAY DE VARIOS AXES

```python
fig,axes = plt.subplots(2,4)
axes[0,0].plot(x,np.cos(x))
axes[0,1].plot(x,np.sin(x),'yellow')
axes[0,2].plot(x,np.tan(x),'r')
axes[0,3].plot(x,np.cos(x)**2,'g')
axes[1,0].plot(x,np.cos(y),'b')
axes[1,1].plot(x,np.sin(y),'purple')
axes[1,2].plot(x,np.tan(y),'orange')
axes[1,3].plot(x,np.cos(y)**2)


fig.tight_layout()
plt.show()
```
