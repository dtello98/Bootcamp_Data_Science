## ðŸ”³ Â¿QuÃ© es un subplot?

Un **subplot** es una tÃ©cnica que permite dibujar **mÃºltiples grÃ¡ficos** dentro de una misma figura en `Matplotlib`.

Esto es Ãºtil cuando queremos comparar diferentes visualizaciones lado a lado o apilar varias grÃ¡ficas en una misma vista.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

# CREAMOS 2 GRAFICOS (SUBPLOTS)

```python
plt.subplot(1,2,1)
plt.plot(x,y,'b--')

plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

# SUBPLOT CON VARIOS GRAFICOS

```python
plt.subplot(1,2,1)
plt.plot(x,y,'b--')
plt.plot(y,x,'r:')

plt.subplot(1,2,2)
plt.pie(x)

plt.show()
```

# CAMBIAMOS ORIENTACIÃ“N DE LOS GRAFICOS

```python
plt.subplot(2,1,1)
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')

plt.subplot(2,1,2)
plt.pie(y)

plt.show()
```
