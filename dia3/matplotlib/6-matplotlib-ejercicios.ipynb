```python
import pandas as pd
df_books = pd.read_csv('https://github.com/cesarmayta/datasets/raw/refs/heads/main/libros.csv')
df_books.head(2)
```

游늷 **Ejercicio 1: Evoluci칩n del precio promedio por a침o**

Muestra c칩mo ha cambiado el precio promedio de los libros en el tiempo.

```python
import matplotlib.pyplot as plt
```

```python
# x = Year
# y = Precio Promedio
```

```python
df_eje1 = df_books.groupby('Year')['Price'].mean().reset_index()
df_eje1
```

```python
x = df_eje1['Year']
y = df_eje1['Price']
```

```python
# grafico
plt.plot(x,y,color='green',label='Precio promedio de libros por a침o')
plt.title('Evoluci칩n del precio promedio de libros por a침o')
plt.xlabel('A침o')
plt.ylabel('Precio Promedio')
plt.legend()
plt.show()
```

游늷 **Ejercicio 2: Comparar calificaci칩n(Rating) promedio por g칠nero**
Compara la calificaci칩n promedio entre libros de ficci칩n y no ficci칩n, mostrar un grafico de barras

```python
df_eje2 = df_books.groupby('Genre')['User Rating'].mean().reset_index()
df_eje2
```

```python
x = df_eje2['Genre']
y = df_eje2['User Rating']
plt.bar(x,y,color=['blue','red'],label='comparaci칩n por genero')
plt.title('Calificaci칩n promedio por Genero')
plt.xlabel('Genero')
plt.ylabel('Calificaci칩n Promedio')
plt.legend()
plt.show()
```
