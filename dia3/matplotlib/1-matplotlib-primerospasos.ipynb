# IMPORTAMOS EL PAQUETE PYPLOT

````python
import matplotlib.pyplot as plt
````

# GENERAMOS DATOS PARA NUESTRO PRIMER PLOT

````python
import numpy as np
x = np.linspace(0,5,11)
x
````

````python
y = x**2
y
````

# CREAMOS NUESTRO PRIMER PLOT

````python
plt.plot(x,y)
plt.show()
````

# CAMBIAMOS EL COLOR DEL GRAFICO

````python
plt.plot(x,y,'r')
plt.show()
````

````python
plt.plot(x,y,'go')
plt.show()
````

````python
plt.plot(x,y,'bo-')
plt.show()
````

````python
plt.plot(x,y,'gD:')
plt.show()
````

````python
plt.plot(x,y,color='#ba57b0',linestyle='--',marker='o')
````

### 游꿛 Par치metros

- **`linestyle`**: `'-'`, `'--'`, `':'`, `'-.`
- **`marker`**: `'o'`, `'s'`, `'*'`, `'x'`, `'^'`
