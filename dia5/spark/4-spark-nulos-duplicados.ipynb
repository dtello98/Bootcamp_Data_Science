# ðŸ§¼ Tratamiento de Nulos y Duplicados en PySpark

```python
import seaborn as sns
df = sns.load_dataset('titanic')
df.head(2)
```

# convertimos el DataFrame de pandas a DataFrame de pyspark

```python
from pyspark.sql import SparkSession

# Inicializar SparkSession
spark = SparkSession.builder.appName("TitanicAnalysis").getOrCreate()

df = spark.createDataFrame(df)

df.printSchema()
df.show(5)
```

```python
df.count()
```

# DROPNA()

```python
df.dropna(how='any').count()
```

```python
df.count() - df.dropna(how='any').count()
```

```python
df.dropna(how='all').count()
```

# dropduplicates

```python
df.dropDuplicates().count()
```
