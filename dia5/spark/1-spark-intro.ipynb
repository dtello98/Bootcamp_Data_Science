# IMPORTAMOS PYSPARK

```python
from pyspark.sql import SparkSession
```

# CREAMOS UNA SESSION A SPARK

```python
spark = SparkSession.builder.appName('PrimeraSesion').getOrCreate()
```

# CREAMOS UN SPARK DATAFRAME

```python
# Crear un DataFrame de ejemplo y realizar operaciones
data = [("Juan", 30), ("Ana", 25), ("Luis", 35)]
columns = ["Nombre", "Edad"]
```

```python
df = spark.createDataFrame(data,columns)
df.show()
```
