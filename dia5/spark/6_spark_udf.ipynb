# Funciones Definidas por el Usuario (UDF) en PySpark

##  Cap铆tulo 1: 驴Qu茅 son las UDF?

###  Definici贸n

Una **UDF (User Defined Function)** es una funci贸n que t煤 defines en Python (o Scala/Java) y que puedes usar dentro de PySpark para aplicar l贸gica personalizada a las columnas de un DataFrame.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType
```

```python
# Crear SparkSession
spark = SparkSession.builder.appName("UDFLibros").getOrCreate()

# Leer archivo libros.csv (sube el archivo en la barra lateral de Colab)
df = spark.read.option("header", "true").option("inferSchema", "true").csv("libros.csv")
df.show(5)
```

# Definir una funci贸n en Python

```python
def clasificar_rating(rating):
    if rating >= 4.5:
        return "Excelente"
    elif rating >= 4.0:
        return "Bueno"
    else:
        return "Regular"
```

# Registrar la UDF en PySpark

```python
clasificar_rating_udf = udf(clasificar_rating, StringType())
```

# Aplicar la UDF sobre el DataFrame

```python
df_resultado = df.withColumn("Clasificaci贸n", clasificar_rating_udf(df["User Rating"]))
df_resultado.select("Name", "User Rating", "Clasificaci贸n").show(10)
```
