```python
import numpy as np
```

```python
temperaturas = np.random.uniform(15,40,30)
temperaturas
```

```python
matriz_temp = temperaturas.reshape(5,6)
matriz_temp
```

# Encuentra:
# - La temperatura máxima por semana
# - El día (índice) en que ocurrió la temperatura máxima por semana
# - La semana con mayor temperatura global

# - La temperatura máxima por semana

```python
max_semana = np.max(matriz_temp,axis=1)
max_semana
```

# - El día (índice) en que ocurrió la temperatura máxima por semana

```python
indice_max_semana = np.argmax(matriz_temp,axis=1)
indice_max_semana + 1
```

# - La semana con mayor temperatura global

```python
semana_max_global = np.argmax(max_semana)
semana_max_global + 1
```
