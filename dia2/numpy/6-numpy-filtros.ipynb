# FILTRADO DE DATOS EN NUMPY

```python
import numpy as np
```

```python
arr = np.linspace(1,10,10,dtype='int64')
arr
```

# SLICING

```python
arr[1:6]
```

```python
arr[2:]
```

```python
arr[:]
```

```python
arr[-1]
```

# CONDICIONALES

```python
condicion = arr > 5
condicion
```

```python
arr[condicion]
```

```python
arr[arr > 5]
```

```python
arr[(arr > 5) & (arr < 9)]
```

```python
arr[arr > 5] = 99
arr
```

# COPY

```python
arr = np.linspace(1,10,10,dtype='int64')
arr
```

```python
arr_copy = arr.copy()
arr_copy
```

```python
arr_copy[:] = 100
arr_copy
```

```python
arr
```

# FILTROS POR AXIS(EJE)
## AXIS = 0 : COLUMNAS
## AXIS = 1 : FILAS

```python
matriz = np.arange(1,21).reshape(5,4)
matriz
```

# SUMATORIA POR FILAS

```python
suma_filas = np.sum(matriz,axis=1)
suma_filas
```

```python
suma_columnas = np.sum(matriz,axis=0)
suma_columnas
```
