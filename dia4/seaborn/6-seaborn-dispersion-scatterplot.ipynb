```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
tips
```

## üîµ ¬øQu√© es `scatterplot` en Seaborn?

`scatterplot` es una funci√≥n de Seaborn que se utiliza para crear **gr√°ficos de dispersi√≥n** (tambi√©n llamados **gr√°ficos de puntos**), mostrando la **relaci√≥n entre dos variables num√©ricas**.

Cada punto en el gr√°fico representa una observaci√≥n en el conjunto de datos.

---

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
plt.show()
```

# RELACI√ìN DENTRE EL TOTAL DE LA CUENTA Y LA PROPINA

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size')
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", palette="pastel")
plt.title("Propinas por d√≠a con paleta pastel")
plt.show()
```

```python
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    style="smoker",
    size="size",
    alpha=0.6,
    palette="muted"
)
plt.title("Propina vs Total con m√∫ltiples dimensiones")
plt.show()
```

### ‚öôÔ∏è Principales par√°metros de `sns.scatterplot`

- `x`:  
  Variable num√©rica que se asigna al eje **horizontal** (eje X).

- `y`:  
  Variable num√©rica que se asigna al eje **vertical** (eje Y).

- `hue`:  
  Variable categ√≥rica o num√©rica que determina el **color** de los puntos.  
  √ötil para diferenciar grupos o categor√≠as.

- `style`:  
  Variable categ√≥rica que define la **forma (s√≠mbolo)** de los puntos, como c√≠rculos, tri√°ngulos, etc.

- `size`:  
  Variable num√©rica que controla el **tama√±o** de cada punto.  
  Sirve para representar una tercera dimensi√≥n num√©rica.

- `alpha`:  
  Nivel de **transparencia** de los puntos (entre `0` y `1`).  
  √ötil cuando hay muchos datos y se superponen.

- `palette`:  
  Especifica la **paleta de colores** a usar para `hue`.  
  Ejemplos: `"deep"`, `"muted"`, `"pastel"`, `"bright"`.

- `data`:  
  El **DataFrame** de Pandas que contiene las variables a graficar.

- `legend`:  
  Controla la leyenda. Puedes usar `"auto"`, `False`, o personalizarla.

- `marker`:  
  Establece manualmente el **s√≠mbolo** de los puntos (por ejemplo `'o'`, `'s'`, `'^'`).
