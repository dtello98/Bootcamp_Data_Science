###  驴Qu茅 es `kdeplot`?

`kdeplot` es una funci贸n de **Seaborn** que permite graficar una **estimaci贸n de la densidad de probabilidad** de una variable num茅rica usando el m茅todo **KDE** (*Kernel Density Estimation* o *Estimaci贸n de Densidad por N煤cleo*).

---

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
tips
```

# KDE BASICO

```python
sns.kdeplot(data=tips,x='total_bill',fill=True)
plt.show()
```

# KDE AGRUPADOS

```python
sns.kdeplot(data=tips,x='total_bill',hue='sex',fill=True)
plt.show()
```
