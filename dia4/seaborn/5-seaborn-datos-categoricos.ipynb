# COUNTPLOT

##  驴Qu茅 es `countplot` en Seaborn?

`countplot` es una funci贸n de **Seaborn** dise帽ada para visualizar **la cantidad (conteo) de observaciones** en cada categor铆a de una variable categ贸rica.

Es 煤til cuando quieres saber **cu谩ntos elementos hay en cada grupo o clase** dentro de una columna categ贸rica.

```python
sns.countplot(data=tips,x='day',hue='sex')
plt.show()
```

# STRIPPLOT

##  驴Qu茅 es `stripplot` en Seaborn?

`stripplot` es una funci贸n de **Seaborn** que se utiliza para **visualizar la distribuci贸n de datos categ贸ricos frente a una variable num茅rica**, colocando **cada observaci贸n individual como un punto** en el gr谩fico.

Es ideal para ver c贸mo se **dispersan los datos** en cada categor铆a.

---

```python
sns.stripplot(data=tips,x='day',y='tip',hue='sex',dodge=True)
plt.show()
```

# CATPLOT

`catplot` (abreviatura de *categorical plot*) es una **funci贸n de alto nivel** en Seaborn que permite crear **gr谩ficos categ贸ricos flexibles y combinables** f谩cilmente.

Es una forma **generalizada** de graficar relaciones entre **variables categ贸ricas** y **variables num茅ricas**, y puede generar varios tipos de gr谩ficos dependiendo del par谩metro `kind`.

---

```python
sns.catplot(data=tips,x='day',col='time',hue='sex',kind='count')
plt.show()
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="box")
plt.suptitle("Total de la cuenta por d铆a (boxplot)", y=1.02)
plt.show()
```

###  Par谩metros principales de `catplot`

- `x`: Variable **categ贸rica**
- `y`: Variable **num茅rica**
- `kind`: Tipo de gr谩fico. Valores posibles:
  - `"strip"`: puntos individuales (por defecto)

  - `"box"`: diagrama de caja
  - `"violin"`: distribuci贸n + box
  - `"bar"`: barra (muestra el promedio)
  - `"point"`: puntos conectados (con l铆neas)
  - `"count"`: conteo de categor铆as (no necesita `y`)
- `hue`, `col`, `row`: Para **crear subgr谩ficos por grupo** y comparar visualmente diferentes categor铆as
