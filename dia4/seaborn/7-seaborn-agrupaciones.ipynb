```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
tips.head(2)
```

## 游댕 쯈u칠 es `jointplot` en Seaborn?

`jointplot` es una funci칩n de alto nivel en Seaborn que permite visualizar **la relaci칩n entre dos variables num칠ricas**, mostrando:

- Un **gr치fico bivariado** en el centro (como scatterplot, kde, reg, etc.)
- Dos **gr치ficos univariados** (histogramas o KDE) en los ejes superior y derecho

Esto lo convierte en una excelente herramienta para **explorar relaciones y distribuciones al mismo tiempo**.

---

### 丘뙖잺 Par치metros principales de `jointplot`

- `x`: Variable para el eje X
- `y`: Variable para el eje Y
- `data`: DataFrame que contiene las columnas
- `kind`: Tipo de gr치fico principal. Puede ser:
  - `"scatter"`: dispersi칩n (por defecto)
  - `"kde"`: densidad (Kernel Density Estimation)
  - `"hist"`: histogramas
  - `"hex"`: hexbin (ideal para datos densos)
  - `"reg"`: regresi칩n lineal
  - `"resid"`: residuales
- `hue`: Variable categ칩rica para diferenciar grupos (desde Seaborn 0.11+)
- `height`: Tama침o del gr치fico (por defecto 6)
- `ratio`: Tama침o relativo de los gr치ficos marginales
- `space`: Espacio entre el gr치fico principal y los marginales

---


```python
sns.jointplot(data=tips,x='total_bill',y='tip',hue='sex',kind='hist')
plt.show()
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='scatter')
```

```python
sns.jointplot(data=tips,x='total_bill',hue='sex',y='tip',kind='kde',fill=True)
```

## 游대 쯈u칠 es `pairplot` en Seaborn?

`pairplot` es una funci칩n de alto nivel en Seaborn que permite crear una **matriz de gr치ficos** donde se muestran las relaciones entre **todas las combinaciones de variables num칠ricas** en un DataFrame.

Para cada par de variables:
- Se genera un **gr치fico de dispersi칩n**.
- En la diagonal se muestra la **distribuci칩n univariada** de cada variable (por defecto, un histograma).

---

### 丘뙖잺 Par치metros principales de `pairplot`

- `data`:  
  DataFrame que contiene los datos.

- `hue`:  
  Variable categ칩rica que permite diferenciar grupos por color.

- `kind`:  
  Tipo de gr치fico bivariado. Valores posibles:
  - `"scatter"`: dispersi칩n (por defecto)
  - `"kde"`: densidad bivariada

- `diag_kind`:  
  Tipo de gr치fico univariado en la diagonal. Opciones:
  - `"hist"`: histograma (por defecto)
  - `"kde"`: densidad univariada

- `palette`:  
  Paleta de colores para `hue`.

- `corner`:  
  Si es `True`, se muestran solo los gr치ficos de la parte inferior de la matriz.

- `plot_kws`:  
  Diccionario con par치metros extra para los gr치ficos (por ejemplo, `s`, `alpha`, etc.)

```python
sns.pairplot(data=tips,hue='sex',palette='coolwarm')
plt.show()
```

```python
sns.pairplot(data=tips,hue='sex',corner=True)
plt.show()
```
