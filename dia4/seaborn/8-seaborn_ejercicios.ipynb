Utiliza `sns.displot` para visualizar la distribuci贸n de la variable num茅rica `age` (edad) de los pasajeros del Titanic.

##  Ejercicio 3: Comparaci贸n categ贸rica con `catplot`

**Enunciado:**  
Usa `sns.catplot` con `kind="count"` para comparar la cantidad de pasajeros seg煤n su `class` (Primera, Segunda o Tercera clase), separados por sexo (`hue="sex"`).
