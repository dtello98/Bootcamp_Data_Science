### 游늵 `displot` en Seaborn

`displot` (abreviatura de *distribution plot*) es una funci칩n de **Seaborn** que se usa para **visualizar distribuciones de datos**.  
Es una funci칩n de **alto nivel** que combina varias capacidades como:

- Histogramas
- Gr치ficos de densidad (`kde`)
- Facetas (m칰ltiples gr치ficos en una sola figura)

Gracias a su flexibilidad, `displot` permite crear visualizaciones complejas de forma sencilla y organizada.

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
tips
```

# CREAMOS GRAFICO CON DISPLOT

```python
sns.displot(data=tips,x='tip')
plt.show()
```

# RELACI칍N ENTRE 2 VARIABLES CON DISPLOT

```python
sns.displot(data=tips,x='total_bill',y='tip')
plt.show()
```

# AGRUPACI칍N DE DATOS

```python
sns.displot(data=tips,x='tip',hue='sex')
plt.show()
```

# GRAFICO DE DENSIDAD CON KDE

```python
sns.displot(data=tips,x='tip',hue='sex',kind='kde')
plt.show()
```

# AGREGAMOS LEYENDAS PROPIAS AL GRAFICO

```python
sns.displot(data=tips,x='tip',hue='sex',kind='kde')
plt.title('Propinas por sexo')
plt.xlabel('Propina')
plt.ylabel('densidad')
plt.show()
```
