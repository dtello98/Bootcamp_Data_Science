### ðŸ“Š Â¿QuÃ© es `histplot`?

`histplot` es una funciÃ³n de **Seaborn** que sirve para crear **histogramas**, es decir, grÃ¡ficos que muestran la **distribuciÃ³n de una variable numÃ©rica** a travÃ©s de **barras** que representan la **frecuencia de valores dentro de ciertos intervalos** (*bins*).

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
tips
```

# CREAMOS NUESTRO PRIMER HISTPLOT

```python
sns.histplot(data=tips,x='total_bill',bins=20)
plt.show()
```

```python
sns.histplot(data=tips,x='total_bill',bins=20,kde=True)
plt.show()
```

#  HISTPLOT CON AGRUPACIONES

```python
sns.histplot(data=tips,x='tip',hue='sex',cumulative=False,stat='count',multiple='dodge')
plt.show()
```
