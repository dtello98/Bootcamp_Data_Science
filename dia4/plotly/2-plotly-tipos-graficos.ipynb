```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('penguins')
df.head(2)
```

```python
import plotly.express as px
```

# GRÁFICO DE DISPERSIÓN (SCATTER)

```python
fig = px.scatter(df,
                 x='flipper_length_mm',
                 y='body_mass_g',
                 color='species',
                 title='Relación entre longitud de aleta y peso')
fig.show()
```

# GRÁFICO DE BARRAS( BAR)

```python
fig = px.bar(df,
             x='species',
             title='cantidad x especie')
fig.show()
```

# HISTOGRAMA (histogram)

```python
fig = px.histogram(df,
                   x='body_mass_g',
                   color='species',
                   title='Distribución de peso por especie')
fig.show()
```

# GRÁFICO DE LINEAS(LINE)

```python
df_sorted = df.sort_values(by=['species','body_mass_g'])
fig = px.line(df_sorted,
              x='species',
              y='body_mass_g',
              color='sex',
              title='Variación de peso por especie')
fig.show()
```

# GRÁFICO DE PIE ( PIE)

```python
fig = px.pie(df,
             names='species',
             title='proporción por especies')
fig.show()
```
