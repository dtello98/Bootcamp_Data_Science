```python
import plotly.express as px
import pandas as pd
import seaborn as sns

df = sns.load_dataset('penguins')
df.head(2)
```

# GRÁFICO DE AREAS

```python
df_sorted = df.sort_values(by='body_mass_g')
fig = px.area(df_sorted,
              x='body_mass_g',
              y='flipper_length_mm',
              color='species',
              title='longitud de aleta x masa corporal')
fig.show()
```

# GRÁFICO DE DENSIDAD

```python
fig = px.density_contour(df,
                         x='body_mass_g',
                         y='flipper_length_mm',
                         color='species',
                         title='Densidad de Masa Corporal vs Longitud de Aleta')
fig.show()
```

# BOX PLOT

```python
fig = px.box(df,
             x='species',
             y='body_mass_g',
             color='species',
             title='Distribución de peso por especie')
fig.show()
```

# GRÁFICOS 3D

```python
fig = px.scatter_3d(df,
                    x='bill_length_mm',
                    y='bill_depth_mm',
                    z='body_mass_g',
                    color='species',
                    title='Relación entre bill length bill depth y body mass')
fig.show()
```
